```python
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import pandas as pd
import time
from tqdm import tqdm
```

```python
import copy
df = pd.read_csv('alba-job-matching-origin.csv')
```

```python
new_df = pd.DataFrame()
new_df["alba_name"] = copy.deepcopy(df["alba_name"])
new_df["sub_code"] = copy.deepcopy(df["sub_code"])

for i in range(1,16):
    new_df["joc_code{0}".format(i)] = copy.deepcopy(df["job_code{0}".format(i)])
new_df.head()
```

```python
new_df.iloc[12,:]
```

```python
import numpy as np
import math
data = pd.DataFrame()
data["alba_name"] = copy.deepcopy(df["alba_name"])
data["sub_code"] = copy.deepcopy(df["sub_code"])
```

```python
driver = webdriver.Chrome('chromedriver.exe')

for idx, row in new_df.iterrows():
    text = ""
    for i in range(2,17):
        if math.isnan(float(row[i])):
            break;
        else:
            try:
                seq = int(row[i])
                driver.get(f"https://www.career.go.kr/cnet/front/base/job/jobView.do?SEQ={seq}")
                time.sleep(0.1)
                lists = driver.find_elements_by_xpath('//*[@id="tab1"]/ul[2]')
                tmp = ""
                for i in lists:
                    tmp = i.text.replace('\n',' ')
                text += tmp
            except Exception as e:
                print(e)
                print(data.loc[idx,'alba_name'] , "fail" ,idx ,"/", len(new_df))
    text.replace('\n'," ")
    data.loc[idx,'context'] = text
    print(data.loc[idx,'alba_name'] , "done" , idx ,"/", len(new_df))
data
```
